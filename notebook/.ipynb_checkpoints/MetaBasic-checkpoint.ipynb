```python
def permittivity2nk(permittivity):
    """ 
    Convert from complex permittivity to (n, k)
    @param permittivity: complex value
    @return: a tuple with n and k
    """
    temp = np.sqrt(permittivity)
    return (temp.real, temp.imag)
```

```python
def nk2permittivity(n, k):
    """ 
    Convert from (n, k) to complex permittivity
    @param n: the real part of complex refractive index
    @param k: the imaginary part of complex refractive index
    @return: complex permittivity
    """
    return (n + 1j * k) ** 2
```

```python
def setResources(fdtd, parallel_job_number, processes, threads, capacity, job_launching_preset):
    """ 
    Set resources in FDTD
    @param fdtd: fdtd object
    @param parallel_job_number: the number of the paralleled jobs
    @param processes: the processes value for each resouce
    @param threads: the threads value for each resouce
    @param capacity: the capacity value for each resouce
    @param job_launching_preset: the job launching preset for each resouce
    """
    # clear all the resources except the first one
    try:
        while True:
            fdtd.deleteresource("FDTD", 2)
    except:
        pass

    # add new resources
    for i in range(parallel_job_number):
        fdtd.addresource("FDTD")
        fdtd.setresource("FDTD", i + 2, "name", "Resource-{}".format(i + 1))
        fdtd.setresource("FDTD", i + 2, "processes", processes)
        fdtd.setresource("FDTD", i + 2, "threads", threads)
        fdtd.setresource("FDTD", i + 2, "capacity", capacity)
        fdtd.setresource("FDTD", i + 2, "Job launching preset", job_launching_preset)

    # clear the first resource
    fdtd.deleteresource("FDTD", 1)
```

```python
def getMatrixCenter(mat):
    """ 
    Return the center value of the matrix
    @param mat: the matrix defined with numpy 2D array
    @return: center value
    """
    w = mat.shape[0]
    l = mat.shape[1]
    if (w % 2 == 0 and l % 2 == 0):
        return (mat[round(w / 2) - 1, round(l / 2) - 1] + mat[round(w / 2) - 1, round(l / 2)] 
            + mat[round(w / 2), round(l / 2) - 1] + mat[round(w / 2), round(l / 2)]) / 4
    elif (w % 2 == 0 and l % 2 != 0):
        return (mat[round(w / 2) - 1, int(np.floor(l / 2))] + mat[round(w / 2), int(np.floor(l / 2))]) / 2
    elif (l % 2 == 0):
        return (mat[int(np.floor(w / 2)), round(l / 2) - 1] + mat[int(np.floor(w / 2)), round(l / 2)]) / 2
    else:
        return (mat[int(np.floor(w / 2)), int(np.floor(l / 2))])
```

```python
def refineMatrix(matrix, multiple):
    """ 
    Return a new matrix with a refinement with the multiple of `multiple`
    @param matrix: the matrix (2D numpy array)
    @param multiple: former resolution = new resolution * multiple
    @return: a new matrix of refined unit (2D numpy array)
    """
    rows, cols = matrix.shape
    rows2, cols2 = rows * multiple, cols * multiple
    matrix2 = np.zeros((rows2, cols2), dtype=matrix.dtype)
    for i in range(rows):
        for j in range(cols):
            matrix2[multiple * i:multiple * i + multiple, multiple * j:multiple * j + multiple] = matrix[i, j]

    return matrix2
```

```python
def integrate(mat, x_vec, y_vec, radius=np.inf):
    """ 
    Return the integrate of `mat` over `x_vec` and `y_vec`
    @param mat: 2D numpy array
    @param x_vec: 1D numpy array
    @param y_vec: 1D numpy array
    @param radius: only the region in the circle (centered at origin) withe the radius of `radius` will be integrated (default: np.inf)
    @return: the integrate
    """
    assert mat.shape[0] == len(x_vec) and mat.shape[1] == len(y_vec), "The dimension of the matrix and vectors don't match."

    if radius == np.inf:
        return np.sum([mat[i, j] * (x_vec[i + 1] - x_vec[i]) * (y_vec[j + 1] - y_vec[j]) \
            for i in range(len(x_vec) - 1) for j in range(len(y_vec) - 1)])
    else:
        return np.sum([mat[i, j] * (x_vec[i + 1] - x_vec[i]) * (y_vec[j + 1] - y_vec[j]) \
            for i in range(len(x_vec) - 1) for j in range(len(y_vec) - 1) if x_vec[i] ** 2 + y_vec[j] ** 2 <= radius ** 2])
```

```python
def getPolarizedField(e_x, e_y, polarization_angle=0):
    """ 
    Return the filed with specified polarization angle
    @param e_x: the x component of field, a ND(spacial dimension + wavelength dimension, ...) numpy array
    @param e_y: the y component of field, a ND numpy array
    @param polarization angle: polarization angle [rad]
    @return: the filed with specified polarization angle, a ND numpy array
    """
    return e_x * np.cos(polarization_angle) + e_y * np.sin(polarization_angle)
```

```python
def getPolarization(polarization, theta, phi):
    """ 
    Solve and return the real polarization angle [rad]
    @param polarization: polarization angle when setting source [rad]
    @param theta: polar angle when setting source [rad]
    @param phi: azimuthal angle when setting source [rad]
    @return: real polarization angle [rad]
    """
    return np.arctan2(
        np.sin(phi) * np.cos(theta) * np.cos(polarization) + \
            np.cos(phi) * np.sin(polarization),
        np.cos(phi) * np.cos(theta) * np.cos(polarization) - \
            np.sin(phi) * np.sin(polarization)
    )
```

```python
# sort phase and make dictionary of index and order
phase_lib_sort_vec = sorted(enumerate(phase_lib_vec), key=lambda x:x[1])
index_order_dict = {phase_lib_sort_vec[i][0]: i for i in range(len(phase_lib_sort_vec))}
order_index_dict = {i: phase_lib_sort_vec[i][0] for i in range(len(phase_lib_sort_vec))}
```

```python
def fom(phase_dest, phase, t):
    """ 
    Figure of merit 
    """
    tolerance_phase = np.deg2rad(2.5)  # phase error we can tolerate considering solving accuracy
    lambda_phase = 2e0  # weight for phase item
    lambda_t = 1e0  # weight for transmission item
    return np.abs(phase - phase_dest) * (np.abs(phase - phase_dest) > tolerance_phase) * lambda_phase + \
        (-1) * t * lambda_t
```

```python
# choose unit from library for the surface
units_pos_width_dict = {}  # dictionary for device units (position and (length, width))
units_pos_index_dict = {}  # dictionary for device units (position and index)
for ii in range(units_rows):
    for jj in range(units_cols):
        phase_dest = space_phase_profile_dest[ii, jj]
        matched = False
        tolerance = np.deg2rad(5)
        while not matched:
            index_begin = 0
            index_end = len(phase_lib_sort_vec) - 1
            # special cases
            if phase_lib_sort_vec[index_end][1] <= phase_dest - tolerance:
                index_begin = index_end
            elif phase_lib_sort_vec[index_begin][1] >= phase_dest + tolerance:
                index_end = index_begin
            # genral case
            else:
                while phase_lib_sort_vec[index_begin][1] < phase_dest - tolerance:
                    index_begin += 1
                while phase_lib_sort_vec[index_end][1] > phase_dest + tolerance:
                    index_end -= 1

            if index_end < index_begin:
                # not matched
                tolerance *= 2
            else:
                # range: [index_begin, index_end]
                fom_min = np.inf
                index_lib_best = 0
                for i in range(index_begin, index_end + 1, 1):
                    index = order_index_dict[i]
                    fom_current = fom(phase_dest, phase_lib_vec[index], t_lib_vec[index])
                    if fom_current < fom_min:
                        fom_min = fom_current
                        index_lib_best = index
                # interpolate the nerghbors of the best-matched unit to find a finer solution
                width_0 = width_lib_vec[index_lib_best]
                length_0 = length_lib_vec[index_lib_best]
                length_idx = round((length_0 - width_min) / width_step)
                width_idx = round((width_0 - width_min) / width_step)
                units_pos_index_dict[(ii, jj)] = index_lib_best
                if length_idx == 0 or length_idx == len(length_vec) - 1 and \
                    width_idx == 0 or width_idx == len(width_vec) - 1:
                    units_pos_width_dict[(ii, jj)] = (length_0, width_0)
                elif length_idx == 0 or length_idx == len(length_vec) - 1:
                    # interpolate width
                    width_1 = width_vec[width_idx - 1]
                    width_2 = width_vec[width_idx + 1]
                    phase_0 = phase_lib_vec[length_idx * len(length_vec) + width_idx]
                    phase_1 = phase_lib_vec[length_idx * len(length_vec) + width_idx - 1]
                    phase_2 = phase_lib_vec[length_idx * len(length_vec) + width_idx + 1]
                    width_11 = (width_1 - width_0) * (phase_dest - phase_0) / (phase_1 - phase_0) + width_0
                    width_22 = (width_2 - width_0) * (phase_dest - phase_0) / (phase_2 - phase_0) + width_0
                    units_pos_width_dict[(ii, jj)] = (length_0, (width_11 + width_22) / 2)
                elif width_idx == 0 or width_idx == len(width_vec) - 1:
                    # interpolate length
                    length_1 = length_vec[length_idx - 1]
                    length_2 = length_vec[length_idx + 1]
                    phase_0 = phase_lib_vec[length_idx * len(length_vec) + width_idx]
                    phase_1 = phase_lib_vec[(length_idx - 1) * len(length_vec) + width_idx]
                    phase_2 = phase_lib_vec[(length_idx + 1) * len(length_vec) + width_idx]
                    length_11 = (length_1 - length_0) * (phase_dest - phase_0) / (phase_1 - phase_0) + length_0
                    length_22 = (length_2 - length_0) * (phase_dest - phase_0) / (phase_2 - phase_0) + length_0 
                    units_pos_width_dict[(ii, jj)] = ((length_11 + length_22) / 2, width_0)
                else:
                    # interpolate both length and width
                    width_1 = width_vec[width_idx - 1]
                    width_2 = width_vec[width_idx + 1]
                    phase_0 = phase_lib_vec[length_idx * len(length_vec) + width_idx]
                    phase_1 = phase_lib_vec[length_idx * len(length_vec) + width_idx - 1]
                    phase_2 = phase_lib_vec[length_idx * len(length_vec) + width_idx + 1]
                    width_11 = (width_1 - width_0) * (phase_dest - phase_0) / (phase_1 - phase_0) + width_0
                    width_22 = (width_2 - width_0) * (phase_dest - phase_0) / (phase_2 - phase_0) + width_0
                    length_1 = length_vec[length_idx - 1]
                    length_2 = length_vec[length_idx + 1]
                    # phase_0 = phase_lib_vec[length_idx * len(length_vec) + width_idx]
                    phase_1 = phase_lib_vec[(length_idx - 1) * len(length_vec) + width_idx]
                    phase_2 = phase_lib_vec[(length_idx + 1) * len(length_vec) + width_idx]
                    length_11 = (length_1 - length_0) * (phase_dest - phase_0) / (phase_1 - phase_0) + length_0
                    length_22 = (length_2 - length_0) * (phase_dest - phase_0) / (phase_2 - phase_0) + length_0
                    units_pos_width_dict[(ii, jj)] = ((length_11 + length_22) / 2, (width_11 + width_22) / 2)
                matched = True
```

```python
# arrangement result
e_arrange_pol_mat = np.zeros((units_rows, units_cols), dtype=np.complex_)
for pos in units_pos_index_dict:
    e_arrange_pol_mat[pos[0], pos[1]] = t_lib_vec[units_pos_index_dict[pos]] * \
        np.exp(1j * (phase_lib_vec[units_pos_index_dict[pos]]))
```
